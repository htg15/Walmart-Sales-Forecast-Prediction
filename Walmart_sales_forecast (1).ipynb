# Data import

```python
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('Walmart.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.sort_values(by='Unemployment')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data['dayOfWeek'] = data['Date'].dt.day_name()
data['month']= data['Date'].dt.month
data['year']= data['Date'].dt.year
```

```python
data.head()
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="year" , y="Weekly_Sales", data=data)
```

```python
sns.boxplot(x="dayOfWeek" , y="Weekly_Sales", data=data)
```

```python
sns.boxplot(x="month" , y="Weekly_Sales", data=data)
```

```python
sns.boxplot(x="Holiday_Flag" , y="Weekly_Sales", data=data)
```

```python
friday_data=data[data['dayOfWeek']=='Friday']
friday_data
```

```python
friday_data=friday_data[friday_data['year']!=2012]
```

```python
friday_data
```

```python
friday_data=friday_data.sort_values('Date')
```

```python
friday_data
```

```python
sns.lineplot(x = 'Date', y = 'Weekly_Sales', hue='year',data = friday_data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))
sns.lineplot(x = 'Date', y = 'Weekly_Sales', hue='year',data = friday_data)
```
